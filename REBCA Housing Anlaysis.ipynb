```python
def plot_box(data, label):
    '''
    INPUT:
    data - (object) Continuous array of ordinal data 
    label - (str) Label for y-axis and title
    
    OUTPUT:
    plt - Box and Whisker plot of the feature 

    Description:
    Returns a Box and Whisker plot of the data and passes a visualisation as well as the mean and standard deviation. 
    '''
    # Calculating spread measures
    median = np.quantile(data[i], 0.5)
    lower_quart = np.quantile(data[i], 0.25)
    upper_quart = np.quantile(data[i], 0.75)
    IQR = upper_quart - lower_quart
    upper_fence = upper_quart + 1.5*IQR
    lower_fence = lower_quart - 1.5*IQR
    
    # Creating plotting object 
    sns.boxplot(y=i,
            orient="h", data= data,)
    plt.grid(True)
    plt.title('Box and whisker plot of {}:'.format(label))
    plt.xlabel('{}'.format(label));

    # Adding the markers
    plt.axvline(median, label = 'Median = {}'.format(median), color = 'black')
    plt.axvline(lower_quart, label = 'Lower Quartile = {}'.format(lower_quart), color = 'purple')
    plt.axvline(upper_quart, label = 'Upper Quartile = {}'.format(upper_quart) , color = 'blue')
    plt.axvline(upper_fence, label = 'Upper Inner Fence = {}'.format(upper_fence), color = 'orange')
    plt.axvline(lower_fence, label = 'Lower Inner Fence = {}'.format(lower_fence), color = 'red')
    
    plt.legend(loc='upper right' , bbox_to_anchor=(1.45, 0.8), shadow=True, ncol=1)

    return plt
```

```python
# 1. Iterating through our columns and printing a boxplot for each feature
# NOTE: Click on the left of the output box to expand the output into the whole notebook
for i in list(private_1_bed_data.columns[1:]):
    fig, ax= plt.subplots()
    fig.set_size_inches(9, 6)
    plt.subplot(1,1,1)
    plot_box(private_1_bed_data, i)
    plt.show()
```

```python
# Function used to create histogram plots
def plot_hist(data, label):
    from scipy.stats import norm
    import scipy 
    '''
    INPUT:
    data - (int) Continuous array of data
    label - (str) Label for x-axis and title
    
    OUTPUT:
    plt - Histogram
    mu - Mean of the array 
    sigma - Standard deviation of the array 
    
    
    
    Description:
    Computes the histogram of the data and passes a visualisation as well as the mean and standard deviation. 
    '''


    # best fit of data
    (mu, sigma) = norm.fit(data)
    
    n, bins, patches = plt.hist(data, 20, density=1, facecolor='green', alpha=0.75)
    y = scipy.stats.norm.pdf( bins, mu, sigma)
    plt.plot(bins, y, 'r--', linewidth=2)
    plt.grid(True)
    plt.axvline(mu, label = 'Mean', color = 'black')
    plt.axvline(np.median(data), label = "Median", color = 'orange')
    plt.xlabel('{}'.format(label))
    plt.ylabel('Frequency')
    plt.title('Histogram of {}: $\mu = {:.3}$, $\sigma = {:.3}$'.format(label, mu, sigma))
    plt.grid(True)
    plt.legend()
              
    return plt, mu, sigma
```

### Normalization and One Hot Encoding

One step I have decided to include is the normalization/scaling of our input features. This will help our model avoid large attribute values from causing any numerical problems during our model training. This [blog](https://medium.com/@urvashilluniya/why-data-normalization-is-necessary-for-machine-learning-models-681b65a05029) and this [paper](chrome-extension://oemmndcbldboiebfnladdacbdfmadadm/https://www.csie.ntu.edu.tw/~cjlin/papers/guide/guide.pdf) help explain the phenomenon.

Applying a scaling to the data does not change the shape of each feature's distribution however, normalization ensures that each feature is treated equally when applying supervised learners. Note that once scaling is applied, observing the data in its raw form will no longer have the same original meaning. Note, we **do not want to transform our target feature**. 

Also, we may have considered using a log-transform on our inputs if we had a distribution where there were a a number of non-trivial outliers. However, in our case, the visualisations indicate this shouldn't be a large worry. We may review removing the outliers as an option to model improvement.

Secondly, I will need to one hot encode the listing_type feature so that we can have a numerical representation of our feature.

```python
# Pipelining the cleaning process from task 3:
def cleaning_data(clean_data, cont_feats):
    
    # Initialize a scaler, then apply it to the features with a default of 0 -> 1
    scaler = MinMaxScaler()
    # copying over our clean dataframe to store in our minmax transformed dataframe
    clean_data_minmax_transform = clean_data.copy()
    # Transforming our new dataframes continuous features
    fitted_scaler = scaler.fit(clean_data[cont_feats[:-1]])
    clean_data_minmax_transform[cont_feats[:-1]] = fitted_scaler.transform(clean_data[cont_feats[:-1]])
    
    # One-hot encode the 'clean_data_minmax_transform' data using pandas.get_dummies()
    final_clean_data = pd.get_dummies(clean_data_minmax_transform)
    
    return final_clean_data


# Pipelining the splitting processes:
def split_data(final_clean_data):
    
    # Randomly selecting 70% of our original dataframe
    final_clean_data_train = final_clean_data.sample(frac = 0.7)
    # Selecting the leftover 30% of the data to transfer into our test set
    final_clean_data_test = final_clean_data.drop(final_clean_data_train.index).copy()
    # Selecting 1/3rd of the test data for the validation set
    final_clean_data_val = final_clean_data_test.sample(frac=0.33333)
    # Removing the validation data points from the test set
    final_clean_data_test.drop(final_clean_data_val.index,inplace=True)
    
    # Resetting the indecies for all datasets
    final_clean_data_train.reset_index(drop=True,inplace=True)
    final_clean_data_test.reset_index(drop=True,inplace=True)
    final_clean_data_val.reset_index(drop=True,inplace=True)
    
    
    # Splitting into target and feature variables for training data
    y_train = final_clean_data_train['monthly_rent'].copy()
    X_train = final_clean_data_train.drop('monthly_rent', axis=1).copy()

    # Splitting into target and feature variables for testing data
    y_test = final_clean_data_test['monthly_rent'].copy()
    X_test = final_clean_data_test.drop('monthly_rent', axis=1).copy()

    # Splitting into target and feature variables for validation data
    y_val = final_clean_data_val['monthly_rent'].copy()
    X_val = final_clean_data_val.drop('monthly_rent', axis=1).copy()
    
    return y_train,X_train,y_test,X_test,y_val,X_val
    
```

## Neural Network


```python
# Setting up the architecture
NN_model = Sequential()

# The Input Layer :
NN_model.add(Dense(128, kernel_initializer='normal',input_dim = X_train.shape[1], activation='relu'))

# The Hidden Layers :
NN_model.add(Dense(256, kernel_initializer='normal',activation='relu'))
NN_model.add(Dense(256, kernel_initializer='normal',activation='relu'))
NN_model.add(Dense(256, kernel_initializer='normal',activation='relu'))

# The Output Layer :
NN_model.add(Dense(1, kernel_initializer='normal',activation='linear'))

# Compile the network :
NN_model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['mean_absolute_error'])
NN_model.summary()
```
